## 1.1 기획의 정의

### 기획이란?

- **기획(Planning)**은 다음의 과정이다:
1. **바람직한 미래 상태**(목표)를 설정하고
2. 이를 달성하기 위한 **행동 계획**을 설계하며
3. **자원**을 체계적으로 배분하는 것

### 어원으로 보는 기획

| 언어 | 어원 | 의미 |
| --- | --- | --- |
| 영어 Planning | 라틴어 "planta" | 건물을 짓기 전에 설계도를 그리다 |
| 한자 企劃 | 企: 발돋움하다, 劃: 칼로 긋다 | 멀리 내다보며 경로를 그리다 |

**핵심 통찰:** 기획 = 멀리 내다보기 + 거기까지 가는 길 그리기

## 1.2 기획 vs 분석 vs 계획 vs 전략

### 핵심 구분

| 개념 | 정의 | 초점 | 핵심 질문 | 산출물 |
| --- | --- | --- | --- | --- |
| **분석(Analysis)** | 현상을 이해하는 것 | 과거/현재 | 무슨 일이 왜 일어났나? | 보고서, 인사이트 |
| **기획(Planning)** | 행동 계획을 설계하는 것 | 미래 | 무엇을 어떻게 할 것인가? | 계획서 |
| **계획(Plan)** | 기획의 문서화된 결과물 | 미래 | - | 문서 |
| **전략(Strategy)** | 장기적 방향 설정 | 미래 | 어디로 갈 것인가? | 비전, 방향 |

### 시각적 비교

````python
# 시각화: 개념 비교

concepts = ['Analysis', 'Planning', 'Plan', 'Strategy']
time_focus = [1, 3, 3, 4]  # 1=Past, 2=Present, 3=Near Future, 4=Far Future
action_level = [1, 4, 3, 2]  # 1=Understanding, 2=Direction, 3=Document, 4=Design
scope = [2, 3, 2, 4]  # 1=Narrow, 4=Broad

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=[time_focus[0], action_level[0], scope[0]],
    theta=['Time Focus', 'Action Level', 'Scope'],
    fill='toself',
    name='Analysis'
))

fig.add_trace(go.Scatterpolar(
    r=[time_focus[1], action_level[1], scope[1]],
    theta=['Time Focus', 'Action Level', 'Scope'],
    fill='toself',
    name='Planning'
))

fig.add_trace(go.Scatterpolar(
    r=[time_focus[3], action_level[3], scope[3]],
    theta=['Time Focus', 'Action Level', 'Scope'],
    fill='toself',
    name='Strategy'
))

fig.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, 5])),
    title='Comparison: Analysis vs Planning vs Strategy',
    height=450
)

fig.show()

print("핵심 인사이트:")
print("  - 분석(Analysis): 과거에 초점, 이해 중심")
print("  - 기획(Planning): 미래에 초점, 행동 설계 중심")
print("  - 전략(Strategy): 먼 미래에 초점, 방향 설정 중심")
````

### Mintzberg의 통찰

> **"기획은 분석(ANALYSIS)에 관한 것이고, 전략은 종합(SYNTHESIS)에 관한 것이다."**
> 
> - 분석: 전체를 부분으로 나누는 것
> - 종합: 부분들을 새로운 전체로 결합하는 것

이것이 "전략적 기획(Strategic Planning)"이 **모순어법(oxymoron)**이 될 수 있는 이유다!

## 1.3 기획의 유형

### 세 가지 주요 유형

## 1.5 전통적 프레임워크가 충분하지 않은 이유

### 일반적인 프레임워크

| 프레임워크 | 목적 | 한계 |
| --- | --- | --- |
| SWOT | 내부/외부 분석 | 정적, 주관적 |
| PEST | 거시 환경 스캔 | 서술적, 인과관계 부재 |
| 5 Forces | 산업 경쟁 분석 | 안정적 구조 가정 |
| BCG Matrix | 포트폴리오 분석 | 지나치게 단순화 |

### 문제점

전통적 프레임워크는:

1. **인과관계를 확립하지 못한다** - 무엇이 무엇을 야기하는가?
2. **불확실성을 무시한다** - 가정이 틀리면?
3. **정적이다** - 변화하는 조건에 적응하지 못한다

### 해결책: 현대적 접근법

| 접근법 | 무엇을 추가하는가 | 해당 장 |
| --- | --- | --- |
| **인과추론** | "X가 정말로 Y를 야기하는가?" | 5-6장 |
| **베이지안 사고** | "우리는 얼마나 확실한가?" | 9장 |
| **시스템 다이내믹스** | "것들이 어떻게 상호작용하는가?" | 6장 |
| **시나리오 기획** | "만약에...?" | 10장 |
| **몬테카를로 시뮬레이션** | "결과의 분포는 어떤가?" | 11장 |

````python
# 시각화: 전통적 vs 현대적 접근법

approaches = pd.DataFrame({
    'Dimension': ['Causality', 'Uncertainty', 'Dynamics', 'Quantification', 'Adaptability'],
    'Traditional (SWOT, PEST)': [1, 1, 1, 2, 1],
    'Modern (Causal, Bayesian)': [5, 5, 4, 5, 4]
})

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=approaches['Traditional (SWOT, PEST)'],
    theta=approaches['Dimension'],
    fill='toself',
    name='Traditional',
    line=dict(color='gray')
))

fig.add_trace(go.Scatterpolar(
    r=approaches['Modern (Causal, Bayesian)'],
    theta=approaches['Dimension'],
    fill='toself',
    name='Modern',
    line=dict(color='blue')
))

fig.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, 6])),
    title='Traditional vs Modern Planning Approaches',
    height=450
)

fig.show()

print("현대적 접근법이 중요한 이유:")
print("  1. 인과성: SWOT은 강점이 성공을 야기하는지 말해주지 않는다")
print("  2. 불확실성: 점추정은 위험을 숨긴다 (9장 참고)")
print("  3. 역동성: 정적 분석은 피드백 루프를 놓친다")
````

## 요약

### 핵심 요점

1. **기획** = 미래 목표 설정 + 행동 계획 설계
