## 제출 방법

1. 5문제 중 **본인이 선택한 최소 3문제** 작성합니다.
2. 각 노트북의 `✏️` 셀에 코드 또는 서술을 작성합니다.
3. **해석(✏️ 해석) 셀은 반드시 본인이 직접 타이핑**하여 작성합니다.
   - AI가 생성한 문장을 그대로 복사·붙여넣기하면 **0점 처리**합니다.
   - 본인의 언어로 분석 결과를 설명해야 합니다.
4. 상단 메뉴에서 **"모두 실행"** 을 클릭하여 전체 셀을 실행합니다.
5. Outline 옆 **⋯ (점 세 개)** → **Export** → **PDF** 를 클릭하여 PDF로 저장합니다.
6. 최소 3개 **PDF 파일**을 제출합니다.
