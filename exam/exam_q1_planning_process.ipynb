## 과제 2. 6단계 기획 프로세스 적용 (8점)

### 요구사항

   - (a) 월별 해지율 추이 라인 차트
   - (b) 불만 유형 비율 파이 차트

2. **(5점)** 6단계 프로세스 중 **1~4단계**를 아래 양식에 작성하세요:

````python
# ✏️ 과제 2-1: 시각화
# (a) 월별 해지율 라인 차트
fig_line = go.Figure()
fig_line.add_trace(go.Scatter(
    x=monthly['월'],
    y=monthly['해지율_pct'],
    mode='lines+markers',
    name='해지율(%)',
    line=dict(color='black', width=2),
    marker=dict(size=8, color='white', line=dict(color='black', width=2))
))

fig_line.update_layout(
    title='월별 구독 해지율 추이',
    xaxis_title='월',
    yaxis_title='해지율 (%)',
    template='plotly_white',
    font=dict(color='black'),
    hovermode='x unified'
)
fig_line.show()

# (b) 불만 유형 비율 파이 차트
fig_pie = go.Figure(data=[go.Pie(
    labels=complaint['불만유형'],
    values=complaint['비율_pct'],
    marker=dict(line=dict(color='black', width=2)),
    textfont=dict(color='black')
)])

fig_pie.update_layout(
    title='고객 불만 유형별 비율',
    font=dict(color='black')
)
fig_pie.show()

print("✓ 시각화 완료: 해지율 추이와 불만 유형 분포를 확인했습니다.")
````

## 과제 3. 간단한 시뮬레이션 (6점)

현재 상태(구독자 3,200명, 해지율 28%, 신규 120명/월)로 **6개월 후 구독자 수**를 계산하세요.

### 요구사항

1. **(3점)** 아래 두 시나리오의 6개월간 구독자 수 변화를 계산하세요:
   - 시나리오 A: 현상 유지 (해지율 28%)
   - 시나리오 B: 배송 개선으로 해지율 20%로 감소

2. **(3점)** 두 시나리오의 구독자 수 추이를 **라인 차트**로 비교하고, 6개월 후 차이를 `print()`로 출력하세요.

````python
# ✏️ 과제 3 풀이
# 시나리오 A vs B 시뮬레이션 + 라인 차트 + 결론 print()

months = [1, 2, 3, 4, 5, 6]
start_customers = 3200
new_customers = 120

# 계산 함수

def simulate_customers(start, churn_rate, new_monthly, months):
    customers = start
    history = []
    for _ in months:
        churn = customers * churn_rate
        customers = customers - churn + new_monthly
        history.append(customers)
    return history

scenario_a = simulate_customers(start_customers, 0.28, new_customers, months)
scenario_b = simulate_customers(start_customers, 0.20, new_customers, months)

fig_sim = go.Figure()
fig_sim.add_trace(go.Scatter(
    x=months,
    y=scenario_a,
    mode='lines+markers',
    name='시나리오 A: 현상 유지 (28%)',
    line=dict(color='black', width=2),
    marker=dict(size=8, color='white', line=dict(color='black', width=2))
))
fig_sim.add_trace(go.Scatter(
    x=months,
    y=scenario_b,
    mode='lines+markers',
    name='시나리오 B: 배송 개선 (20%)',
    line=dict(color='gray', width=2, dash='dash'),
    marker=dict(size=8, color='white', line=dict(color='black', width=2))
))

fig_sim.update_layout(
    title='6개월 후 구독자 수 시나리오 비교',
    xaxis_title='개월',
    yaxis_title='구독자 수',
    template='plotly_white',
    font=dict(color='black'),
    hovermode='x unified'
)
fig_sim.show()

final_a = scenario_a[-1]
final_b = scenario_b[-1]
difference = final_b - final_a

print(f"시나리오 A 6개월 후 구독자 수: {int(final_a)}명")
print(f"시나리오 B 6개월 후 구독자 수: {int(final_b)}명")
print(f"6개월 후 시나리오 B가 시나리오 A보다 {int(difference)}명 더 유지됩니다.")
````
