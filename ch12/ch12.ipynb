## 1. 전통적 기획의 한계

전통적 기획은 다음 가정 위에 서 있다.

- 환경이 안정되어 있다.
- 1년 전 예측이 1년 후에도 유효하다.
- 계획대로 실행하면 결과가 나온다.

**현실**

> "No plan survives first contact with the enemy." — Helmuth von Moltke

신사업·신정책·격변기에는 계획대로 가는 일이 거의 없다. 적응적 기획은 "바뀔 것"을 가정한다.

## 2. 적응적 기획 사이클

```
   Plan (계획)
      v
   Do (실행)
      v
   Check (모니터링)  ← 11주차 KPI
      v
   Adapt (계획 수정) ← 트리거 발동 시
      v
   (다시 Plan)
```

핵심은 **Adapt 단계가 정해진 시점에 정해진 트리거로 자동 발동**되어야 한다는 점이다. 그렇지 않으면 "항상 계획대로"가 되어 변경이 영원히 미뤄진다.

| 구분 | 전통적 기획 | 적응적 기획 |
|------|-----------|-----------|
| 가정 | 환경이 안정 | 환경이 변한다 |
| 주기 | 1년에 한 번 | 분기·반기마다 갱신 |
| 변경 | 어렵다 | 트리거에 따라 자동 |
| 학습 | 사후 평가 | 진행 중 학습 |

## 3. 롤링 플래닝 매뉴얼

전통적 기획은 1년 계획을 1년 동안 그대로 간다. 롤링 플래닝은 **시간 지평을 4구간으로 나눠** 각각 다른 상세도와 갱신 주기를 둔다.

| 기간 | 상세도 | 갱신 주기 |
|------|------|---------|
| 다음 분기 | 매우 상세 (주별 일정) | 월 |
| 6개월~1년 | 보통 (월별) | 분기 |
| 1~3년 | 개략 (분기별) | 반기 |
| 3~5년 | 비전·방향만 | 연 |

**작성 5단계**

1. 시간 지평 4구간을 정한다.
2. 구간별 상세도를 정한다.
3. 각 구간의 갱신 주기를 달력에 못 박는다.
4. 갱신 시 가장 가까운 구간만 다시 짜고, 먼 구간은 비전만 점검한다.
5. 계획 수정 이력을 한 페이지로 남긴다 (왜 바뀌었는가).

## 4. 계획 수정 트리거

트리거는 "이게 발생하면 계획을 다시 본다"는 사전 합의다. 11주차 KPI 시스템과 직접 연결된다.

**유형별 예시**

| 유형 | 예시 |
|------|------|
| KPI 임계값 위반 | 분기 매출 -20%, 신청 건수 -30% |
| 외부 이벤트 | 신규 규제, 경쟁사 인수합병, 기후 재해 |
| 정해진 일정 | 분기 종료, 시범사업 종료 |
| 가설 검증 결과 | 베이지안 사후확률이 임계 이하 |

**작성 매뉴얼**

1. **핵심 가정 5개**를 적는다 ("X가 Y이면 계획이 통한다").
2. 각 가정마다 **정상 범위**와 **이탈 조건**을 적는다.
3. 트리거 발동 시 **누가 무엇을 결정**하는지 적는다 (RACI).
4. 트리거를 KPI 대시보드에 표시한다.

````python
triggers = pd.DataFrame([
    ["A1", "Pilot demand strong",  "Apply > 200/quarter",     "Apply < 100 for 2 quarters",        "Director", "Reduce target / extend timeline"],
    ["A2", "Onboarding works",     "Completion >= 80% in 30d","Completion < 60% for 1 quarter",    "Manager",  "Redesign onboarding"],
    ["A3", "Cost on plan",         "Spend +-5% of plan",      "Spend > +15% YTD",                  "CFO",      "Cut scope or request budget"],
    ["A4", "Outcome rate holds",   "Employment >= 65% @12mo", "Employment < 50% @12mo",            "Director", "Pivot cohort or industry"],
    ["A5", "Stable regulation",    "No new rule",             "Eligibility-altering rule passed",  "Director", "Pause and reassess"],
], columns=["id", "assumption", "in_band", "trigger", "owner", "next_action"])
triggers
````

````python
# 트리거 평가 시뮬레이션 — 현재 어느 트리거가 깨졌는가
current_status = {
    "A1": False,  # 신청 양호
    "A2": True,   # 등록률 저조
    "A3": False,  # 예산 정상
    "A4": False,  # 아직 측정 시점 아님
    "A5": False,  # 신규 규제 없음
}

evaluation = triggers.copy()
evaluation["breached?"] = evaluation["id"].map(lambda i: "YES" if current_status[i] else "no")
evaluation["action_now"] = evaluation.apply(
    lambda r: r["next_action"] if current_status[r["id"]] else "", axis=1
)
evaluation[["id", "assumption", "trigger", "breached?", "owner", "action_now"]]
````

## 5. After Action Review (AAR)

AAR은 미군에서 시작된 학습 도구로, 4개 질문으로 빠르게 점검한다.

**규칙**

- 회의 30분 이내.
- 비난 금지 — 시스템·프로세스에 초점.
- 발견 사항은 문서화하고 공유.

**4개 질문**

1. 무엇이 일어나기를 기대했는가?
2. 실제로 무엇이 일어났는가?
3. 왜 차이가 났는가?
4. 다음에 무엇을 다르게 할 것인가?

````python
aar = pd.DataFrame({
    "question": [
        "1. What did we expect?",
        "2. What actually happened?",
        "3. Why was there a gap?",
        "4. What will we do differently next?",
    ],
    "answer": [
        "Q2 신청 200건, 완료율 80%",
        "신청 230건, 완료율 58%",
        "모집은 성공했으나 등록 절차 복잡 (서류 8종)",
        "등록 양식 단순화, 자동 입력 PoC, 상담 직원 1명 배치",
    ],
})
aar
````

## 6. 통합 기획서 구조 (기말 프로젝트)

한 학기 동안 배운 도구를 **하나의 주제**에 적용한 결과물이다. 각 섹션은 1~2 페이지로 짧게.

| 섹션 | 활용 주차 | 산출물 |
|------|---------|------|
| 1. 문제 정의 (MECE·로직 트리) | 3주차 | Why/How 트리 1장 |
| 2. 인과 분석 (DAG) | 4주차 | DAG 1장 + 핵심 인과 가설 |
| 3. 시스템 사고 (CLD) | 5주차 | 인과 루프 다이어그램 1장 |
| 4. 베이지안 의사결정 | 6주차 | 사전·사후 + 결정 권고 |
| 5. 시나리오 플래닝 | 7주차 | 4시나리오 + 강건한 전략 |
| 6. 몬테카를로 시뮬레이션 | 9주차 | 결과 분포 + 토네이도 |
| 7. 실행 계획과 리스크 | 10주차 | OKR + RACI + 리스크 레지스터 |
| 8. KPI와 모니터링 | 11주차 | KPI 5~7개 + 대시보드 |
| 9. 적응적 기획 | 12주차 | 트리거 5개 + 롤링 플랜 + AAR |
| 10. 통합 시뮬레이션 워크숍 | 13주차 | 통합 사례 분석 결과 |


## 7. 바이브 코딩 프롬프트

**1단계: 트리거 표 생성**

```text
내 정책의 핵심 가정 5개와 KPI 트리거 5개를 표로 정리해줘.
각 트리거마다 발동 조건(in_band, trigger), 책임자, 다음 행동을 적어줘.
```

**2단계: 롤링 플랜 작성**

```text
내 사업의 시간 지평 4구간(분기/반기/연/장기) 롤링 플랜 표를 만들어줘.
각 구간의 상세도와 갱신 주기를 적어줘.
```

**3단계: AAR 자동 채움**

```text
이번 분기 KPI 데이터와 목표를 줄게.
AAR 4개 질문 양식을 만들고, 데이터를 바탕으로 1번과 2번을 자동 채워줘.
3번과 4번은 빈 칸으로 두되, 채울 때 참고할 질문 가이드를 옆에 적어줘.
```

## 오늘의 핵심

적응적 기획은 "계획을 자주 바꾸자"가 아니다. **언제·왜·어떻게 바꿀지를 계획에 미리 적어 두는 것**이다. 트리거가 없으면 변경은 영원히 미뤄지고, 트리거가 너무 자주 발동하면 일관성이 무너진다.
