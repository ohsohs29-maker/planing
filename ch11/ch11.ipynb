## 1. 성과 측정의 본질

피터 드러커: *"측정되지 않으면 관리되지 않는다."*

기획서에 KPI가 없으면 실행 단계에서 진행 상황을 알 수 없다. 그러나 **잘못 측정하면 잘못된 행동을 부른다**. 측정 지표가 행동을 결정하기 때문이다.

사례:

- 콜센터에서 "통화 시간"만 KPI로 잡으면 직원들은 빨리 끊기 시작한다 → 고객 만족도 하락.
- 학교에서 "평균 점수"만 KPI로 잡으면 어려운 학생을 포기하기 쉽다 → 형평성 악화.

**기획자의 질문**

> 우리 KPI가 의도하지 않은 행동을 유발하지 않는가?

## 2. 선행 지표 vs 후행 지표

| 구분 | 선행 지표 (Leading) | 후행 지표 (Lagging) |
|------|-------------------|--------------------|
| 시점 | 결과 전 | 결과 후 |
| 통제 가능성 | 통제 가능 | 통제 불가 |
| 다이어트 | 일일 칼로리 섭취 | 체중 |
| 매출 | 영업 미팅 수 | 분기 매출 |
| 정책 | 시범 신청 건수 | 1년 후 고용률 |

한 줄 요약:

> 후행 지표만 보면 손쓸 시점에는 이미 늦는다. **선행 지표를 함께 두자**.

균형 비율 권장: 선행 50% + 후행 50%.

## 3. SMART 매뉴얼

KPI를 작성할 때 다음 5가지를 점검한다.

| 글자 | 점검 질문 |
|------|---------|
| **S**pecific | 무엇을 측정하는지 한 문장으로 적을 수 있는가 |
| **M**easurable | 숫자로 측정 가능한가 |
| **A**chievable | 달성 가능한 수준인가 |
| **R**elevant | 핵심 목표와 연결되는가 |
| **T**ime-bound | 언제까지인가 |

**나쁜 KPI** → "고객 만족도 향상"

**좋은 KPI** → "2026년 12월까지 NPS 30점 달성 (현재 18점)"

## 4. Balanced Scorecard 4관점

KPI를 한 영역에 몰면 균형이 깨진다. 4관점에 골고루 둔다.

| 관점 | 핵심 질문 | 예시 |
|------|---------|------|
| 재무 | 주주·예산은 만족하는가 | 순이익, ROI |
| 고객 | 시민·고객은 만족하는가 | NPS, 재이용률 |
| 내부 프로세스 | 내부 운영은 효율적인가 | 처리 시간, 오류율 |
| 학습·성장 | 조직은 미래에 준비되어 있는가 | 직원 교육 시간 |

**KPI 카탈로그 예시 — 청년 일자리 정책**

````python
kpis = pd.DataFrame([
    ["Pilot applications",        "Leading", "Customer",          "Apply count > 200/quarter"],
    ["Onboarding completion",     "Leading", "Internal Process",  ">= 80% within 30 days"],
    ["Training hours / staff",    "Leading", "Learning & Growth", ">= 12h/year"],
    ["Spend rate vs plan",        "Leading", "Financial",         "Within +-5% of plan"],
    ["Employment rate (12 mo)",   "Lagging", "Customer",          ">= 65%"],
    ["Cost per outcome",          "Lagging", "Financial",         "<= 8M KRW per job"],
    ["Process error rate",        "Lagging", "Internal Process",  "<= 2%"],
    ["NPS",                       "Lagging", "Customer",          ">= 30"],
], columns=["kpi", "type", "bsc_perspective", "smart_target"])

kpis
````

## 5. 이상 탐지의 직관

복잡한 알고리즘 없이도 두 가지 방법으로 충분히 잡을 수 있다.

1. **이동 평균 (Moving Average)**: 최근 N주 평균. 단기 흐름이 추세에서 어긋나는지 본다.
2. **표준편차 밴드 (±2σ)**: 평균에서 표준편차 2배를 벗어나면 이상 의심.

```
Alert  if  현재 값  >  이동평균 + 2 * 이동표준편차
       or  현재 값  <  이동평균 - 2 * 이동표준편차
Watch  if  현재 값  >  이동평균 + 1 * 이동표준편차
Good   otherwise
```

권장 윈도우: 주간 데이터는 8주, 월간 데이터는 6개월.

**기획자의 주의사항**

- Alert가 자주 뜨면 임계값(σ 배수)을 키운다.
- 계절성(분기별·요일별 패턴)이 있으면 이동 평균만으로는 부족하다 → 같은 분기끼리 비교.

````python
# 시범 사업 지원 신청 건수 (52주)
rng = np.random.default_rng(1)
week = np.arange(52)
trend = 100 + 0.3 * week
noise = rng.normal(0, 4, 52)
series = trend + noise
series[30] += 18   # 30주차 급증 (주변 보조금 정책 영향?)
series[45] -= 22   # 45주차 급감 (이슈 발생?)

df = pd.DataFrame({"week": week, "value": series})
df["ma"] = df["value"].rolling(8, min_periods=1).mean()
df["sd"] = df["value"].rolling(8, min_periods=1).std().fillna(0)
df["upper"] = df["ma"] + 2 * df["sd"]
df["lower"] = df["ma"] - 2 * df["sd"]
df["status"] = np.where(
    (df["value"] > df["upper"]) | (df["value"] < df["lower"]), "Alert",
    np.where(df["value"] > df["ma"] + df["sd"], "Watch", "Good"),
)
df.tail(10)[["week", "value", "ma", "upper", "lower", "status"]]
````

````python
plt.figure(figsize=(9, 4))
plt.plot(df["week"], df["value"], color="black", linewidth=1, label="Value")
plt.plot(df["week"], df["ma"], color="gray", linestyle="--", linewidth=0.8, label="MA(8)")
plt.fill_between(df["week"], df["lower"], df["upper"], color="lightgray", alpha=0.4, label="+-2 sigma")
alerts = df[df["status"] == "Alert"]
plt.scatter(alerts["week"], alerts["value"], s=70, facecolors="white", edgecolors="black", linewidths=1.2, label="Alert")
plt.xlabel("Week", color="black")
plt.ylabel("Application count", color="black")
plt.title("Pilot applications with +-2 sigma band", color="black")
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
````

## 6. KPI 대시보드 매뉴얼

**대시보드 작성 6단계**

1. 핵심 KPI **5~7개**를 고른다 (너무 많으면 안 본다).
2. 각 KPI에 **목표값**과 **임계값**을 설정한다.
3. 색상 대신 **텍스트로 상태**를 표시한다 (Good / Watch / Alert) — 색맹·인쇄·요약문 호환을 위해.
4. 한 화면에 다 보이도록 한다.
5. 임원 보고용은 후행 + 선행 1개씩, 실무용은 선행 위주.
6. 매주 같은 시간에 갱신·검토한다.

````python
# 간단 대시보드: 4개 KPI 한 줄 요약
dashboard = pd.DataFrame([
    {"kpi": "Pilot applications (week 51)", "current": round(df.iloc[-1]["value"], 1),
     "target": 130, "status": df.iloc[-1]["status"]},
    {"kpi": "Onboarding completion (last month)", "current": 0.78, "target": 0.80, "status": "Watch"},
    {"kpi": "Spend rate vs plan (YTD)", "current": 1.04, "target": 1.00, "status": "Good"},
    {"kpi": "Process error rate (last week)", "current": 0.018, "target": 0.020, "status": "Good"},
])
dashboard
````

## 7. 바이브 코딩 프롬프트

오늘 실습에서 AI에게 요청할 문장이다.

**1단계: KPI 추천**

```text
나는 [정책/사업명]의 KPI 체계를 설계하려고 해.
선행 지표 4개와 후행 지표 4개를 추천해줘.
각각을 SMART 형식으로 작성하고, Balanced Scorecard 4관점으로 분류도 해줘.
표 형식으로 정리해줘.
```

**2단계: 시계열 데이터 생성**

```text
한 KPI를 골라 1년치 주별 가짜 데이터를 만들어줘.
추세 + 잡음에 더해서, 30주차와 45주차에는 이상치를 넣어줘.
```

**3단계: 이상 탐지**

```text
위 데이터에 8주 이동 평균과 +-2 시그마 밴드를 그려서 이상 시점을 찾는 Python 코드를 만들어줘.
그래프는 흑백, 이상점은 흰 점에 검은 테두리로 표시해줘.
```

**4단계: 대시보드**

```text
KPI 4~6개의 현재값·목표값·상태(Good/Watch/Alert)를 표로 정리하는 코드를 만들어줘.
임원 보고용으로 한 화면에 다 보이도록 해줘.
```
