이번 시간은 새 도구를 배우지 않는다. 한 학기 동안 쌓은 결과물을 **5분 발표**로 정리하고, 동료의 피드백을 받아 통합 기획서의 마지막 다듬기를 한다.

수업이 끝나면 다음을 할 수 있어야 한다.

2. 동료 발표를 정해진 rubric으로 평가한다.
3. 받은 피드백 중 무엇을 수용하고 무엇을 거절할지 판단한다.
4. 최종 통합 기획서를 제출한다.

> **한 줄 요약**: 발표는 한 학기 학습을 5분으로 압축하는 연습이다. 피어 리뷰는 동료의 시선으로 자기 사고의 사각지대를 보는 거울이다.

## 1. 발표 구성 (5분 + 질의 2분)

| 시간 | 내용 | 슬라이드 수 |
|------|------|-----------|
| 0:30 | 정책 주제와 핵심 질문 | 1 |
| 1:00 | 인과 구조 (DAG) + 시스템 구조 (CLD) | 1 |
| 1:00 | 베이지안 결정 + 시나리오 4개 | 1 |
| 1:00 | 몬테카를로 결과 (분포·달성 확률) | 1 |
| 1:30 | 통합 권고안 + 핵심 가정 + 적응 트리거 | 1 |
| **5:00** | **합계** | **5** |
| 2:00 | 질의응답 | — |

**발표 원칙**

- 5장의 슬라이드 = 5개 도구 결과 = 5개 핵심 메시지
- 한 슬라이드 = 한 메시지 (제목이 곧 메시지)
- 모든 그래프는 흑백, 한글 라벨
- 숫자는 출처와 가정을 함께 표시 (예: "5분위 +8% — 가정: 임대료 상승 ≤ 5%")


---

## 2. 발표 슬라이드 골격

각 슬라이드의 제목과 본문 가이드.

**슬라이드 1: 주제와 핵심 질문**

- 제목: "[정책 주제]에 [예산/자원]을 어떻게 배분할 것인가?"
- 본문: 정책 옵션 3개(A/B/C) + 한 줄 설명

**슬라이드 2: 인과·시스템 구조**

- 제목: "[정책]의 효과는 [숨은 원인]·[피드백 루프]에 의해 달라진다"
- 본문 좌: DAG (변수 5~7개)
- 본문 우: CLD (강화 루프 1개 + 균형 루프 1개)

**슬라이드 3: 베이지안 결정 + 시나리오**

- 제목: "현재 시점 권고: [옵션 X]. 단, 미래에 따라 [옵션 Y]도 가능"
- 본문 좌: 베이지안 사전→사후 변화 표
- 본문 우: 2축 시나리오 매트릭스 + 강건한 전략

**슬라이드 4: 몬테카를로 결과**

- 제목: "5년 후 목표 달성 확률: [N]% — [충분/부족] 평가"
- 본문 좌: 결과 분포 히스토그램
- 본문 우: 토네이도 차트 (가장 영향 큰 변수 3개)

**슬라이드 5: 통합 권고안**

- 제목: "[정책]의 권고: 1년차 A, 2년차 A→B, 3년차 B→C (단계적 조정)"
- 본문: 권고안 3~5줄 + 핵심 가정 1개 + 적응 트리거 1개

## 3. 피어 리뷰 rubric (5점 척도)

각 항목 1~5점. 총 25점 만점. 동료 3명 평가의 평균을 채택한다.

| 항목 | 1점 (낮음) | 3점 (보통) | 5점 (높음) |
|------|-----------|-----------|-----------|
| **1. 문제 정의 명확성** | 주제가 막연하다 | 주제는 명확하나 옵션 비교가 약하다 | 옵션 3개가 분명하고 비교 기준이 있다 |
| **2. 도구 적용 정확성** | 도구 사용이 형식적이다 | 도구별 결과가 있으나 해석이 약하다 | 5개 도구가 각자 다른 통찰을 준다 |
| **3. 결과 통합력** | 도구별 결과가 분리되어 있다 | 통합 권고가 있으나 충돌 지점은 안 보인다 | 충돌 지점이 명시되고 통합 논리가 있다 |
| **4. 가정·한계 인지** | 가정이 명시되지 않는다 | 가정 일부 명시 | 핵심 가정 + 깨졌을 때 대응까지 있다 |
| **5. 발표·시각화** | 슬라이드가 글 위주, 그래프 불명확 | 그래프는 있으나 메시지 약함 | 슬라이드 제목이 곧 메시지, 그래프가 메시지 보조 |

**피어 리뷰 작성 양식**

```text
발표자: [이름]
주제: [정책 주제]

1. 문제 정의 명확성: __점 — [한 줄 코멘트]
2. 도구 적용 정확성: __점 — [한 줄 코멘트]
3. 결과 통합력: __점 — [한 줄 코멘트]
4. 가정·한계 인지: __점 — [한 줄 코멘트]
5. 발표·시각화: __점 — [한 줄 코멘트]

가장 인상 깊은 점: [1줄]
가장 의문스러운 점: [1줄]
이 발표를 듣고 내 발표에서 바꾸고 싶은 점: [1줄]
```

## 4. 자기평가 체크리스트

발표 전 자가 점검. 모두 ✓이면 발표 준비 완료.

**도구 적용**

- [ ] 5개 도구(DAG·CLD·베이지안·시나리오·몬테카를로) 결과가 있다
- [ ] 도구별 결과가 서로 충돌하는 지점을 1개 이상 식별했다
- [ ] 도구별 결과를 한 줄로 요약할 수 있다

**가정·한계**

- [ ] 핵심 가정 1개를 명시했다
- [ ] 그 가정이 깨지면 무엇이 바뀌는지 설명할 수 있다
- [ ] 데이터 출처(공공데이터·문헌 등)가 명시되어 있다

**AI 활용 투명성**

- [ ] AI에게 보낸 핵심 프롬프트 1~2개를 보고서에 첨부했다
- [ ] AI 출력 중 사람이 수정한 부분을 표시했다
- [ ] AI가 생성한 가짜 인물·가짜 통계가 없다는 것을 확인했다

**발표 형식**

- [ ] 슬라이드 5장 (각 1메시지)
- [ ] 모든 그래프 흑백 + 한글 라벨
- [ ] 5분 안에 마칠 수 있는 분량

| 단계 | AI에게 보낸 핵심 프롬프트 | AI 출력 그대로 사용 | 사람이 수정·반박한 점 |
|------|--------------------------|------------------|--------------------|
| 1단계 DAG | | | |
| 2단계 CLD | | | |
| 3단계 베이지안 | | | |
| 4단계 시나리오 | | | |
| 5단계 몬테카를로 | | | |
| 통합 권고 | | | |

**AI 환각 점검 (필수)**

- 이 보고서에 들어간 통계 수치 중 출처를 댈 수 없는 것이 있는가? → 있으면 빨간 표기 또는 삭제

## 6. 발표 후 피드백 통합 워크시트

피어 리뷰 3건을 받은 뒤, 다음 워크시트로 정리한다.

| 받은 피드백 | 동의·거절 | 이유 | 기획서 수정 사항 |
|-----------|---------|------|---------------|
| 예: "DAG에 가구 소득이 빠졌다" | 동의 | 교란변수 누락 — 분석 신뢰 ↓ | DAG에 노드 추가, 분석 재실행 |
| 예: "옵션 C는 너무 보수적" | 거절 | 시나리오 4 (빙하기)에서는 C가 필요 | 거절 이유를 기획서 한계 절에 명시 |
| | | | |
| | | | |
| | | | |

**거절 원칙**

- 받은 피드백을 모두 수용할 필요는 없다.
- 거절할 때는 **기획서에 거절 이유를 명시**한다 ("동료 X가 Y를 지적했으나 시나리오 Z에서는 본 권고가 유효하다").
- 동의할 수도 거절할 수도 없는 경우 = "추가 데이터 필요"로 한계 절에 기록.

## 7. 발표 보조 코드 (선택)

발표 슬라이드용 그래프를 만드는 도우미 코드. 자기 결과로 바꿔 사용한다.


````python
# 발표용 통합 요약 차트 예시
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import numpy as np

# 한글 폰트
for font in ["AppleGothic", "Malgun Gothic", "맑은 고딕", "NanumGothic"]:
    if font in {f.name for f in fm.fontManager.ttflist}:
        plt.rcParams["font.family"] = font
        break
plt.rcParams["axes.unicode_minus"] = False

# 예시: 3개 옵션의 베이지안 사전→사후 + 시나리오별 강건성
options = ["옵션 A", "옵션 B", "옵션 C"]
prior = [0.33, 0.33, 0.34]
posterior = [0.50, 0.30, 0.20]   # 베이지안 결과 (예시)
robust = [0.6, 0.5, 0.8]          # 시나리오 4개에서 작동 비율 (예시)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# 좌: 사전 vs 사후
x = np.arange(len(options))
axes[0].bar(x - 0.2, prior, width=0.4, color="white", edgecolor="black", label="사전")
axes[0].bar(x + 0.2, posterior, width=0.4, color="gray", edgecolor="black", label="사후")
axes[0].set_xticks(x); axes[0].set_xticklabels(options)
axes[0].set_ylabel("확률"); axes[0].set_title("베이지안: 사전 → 사후")
axes[0].legend()

# 우: 시나리오 강건성
axes[1].bar(options, robust, color="white", edgecolor="black")
axes[1].set_ylabel("4개 시나리오 중 작동 비율")
axes[1].set_title("시나리오 강건성")
axes[1].set_ylim(0, 1)

plt.tight_layout()
plt.show()
````
