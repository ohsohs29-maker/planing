## 1. 좋은 전략이 왜 실패하는가

전략 슬라이드는 멋진데 결과가 안 나오는 5가지 패턴.

1. **추상적이다** — 누가 무엇을 할지 모른다.
2. **책임이 흩어진다** — 결정이 미뤄진다.
3. **리스크가 식별되지 않는다** — 사고가 나면 뒤늦게 대응한다.
4. **KPI가 없다** — 진행을 모른다 (11주차).
5. **환경 변화에 따라가지 못한다** — 적응 트리거가 없다 (12주차).

10주차는 1, 2, 3을 다룬다.

**한 줄 요약**

> 실행 계획은 "누가 무엇을 언제 하고 어떤 위험을 어떻게 대비할지" 한 페이지로 답할 수 있어야 한다.

## 2. OKR로 야심찬 목표 작성하기

**OKR = Objective + Key Results**

- **Objective**: 정성적·야심적·기억 가능한 한 문장.
- **Key Results**: 측정 가능한 결과 3~5개.
- 분기마다 새로 짠다.
- 목표 달성도 0.6~0.7이면 성공 (1.0이면 너무 쉬운 목표).

| 도구 | 강점 | 한계 | 적합 상황 |
|------|------|------|---------|
| **OKR** | 야심·정렬 | 분기 단위, 문화 필요 | 분기 목표·신사업 |
| **SMART** | 구체성·검증성 | 야심·정렬 약함 | 운영 KPI |

**나쁜 예** → "청년 일자리를 늘린다" (측정 불가)

**좋은 예** → Objective: "지역 청년의 첫 직장 진입을 빠르고 쉽게 만든다"

- KR1: 분기 신청 200건
- KR2: 12개월 후 취업률 ≥ 65%
- KR3: 1인당 비용 ≤ 8백만원

## 3. WBS로 일 쪼개기

**Work Breakdown Structure** — 큰 일을 작은 일로 쪼갠다. **3 레벨**이 가독성이 좋다.

```
Level 1: 정책 시범사업 운영 (전체)
  Level 2: 모집 / 교육 / 매칭 / 평가
    Level 3 (모집): 홍보 / 신청 시스템 / 서류 검토 / 면담
```

**작성 매뉴얼**

1. Level 1: 사업 전체 (이름 1개).
2. Level 2: 4~6개 큰 흐름 (각 흐름이 8주 이내가 이상적).
3. Level 3: 각 흐름을 3~4개 작업으로 쪼갠다.
4. 각 Level 3 작업에 **기간(주)**, **담당자**, **산출물**을 적는다.

````python
wbs = pd.DataFrame([
    ["1", "Recruit",  "1.1 Outreach campaign",       2, "Comm Lead",  "Campaign assets"],
    ["1", "Recruit",  "1.2 Online application form", 2, "IT Lead",    "Live form"],
    ["1", "Recruit",  "1.3 Document review",         3, "Ops Lead",   "Eligibility list"],
    ["2", "Train",    "2.1 Training curriculum",     3, "Edu Lead",   "Curriculum doc"],
    ["2", "Train",    "2.2 Run cohort",              8, "Edu Lead",   "Trained cohort"],
    ["3", "Match",    "3.1 Employer onboarding",     2, "Biz Lead",   "Partner list"],
    ["3", "Match",    "3.2 Interview matching",      4, "Biz Lead",   "Match results"],
    ["4", "Evaluate", "4.1 6-month survey",          1, "Eval Lead",  "Survey data"],
    ["4", "Evaluate", "4.2 12-month outcome report", 2, "Eval Lead",  "Outcome report"],
], columns=["L1_id", "L1_name", "L3_task", "weeks", "owner", "deliverable"])
wbs
````

## 4. RACI로 책임 분리하기

각 작업에 4역할을 배정한다.

| 글자 | 의미 | 인원 |
|------|------|------|
| **R**esponsible | 직접 일하는 사람 | 1명 이상 |
| **A**ccountable | 결정·승인하는 사람 | **반드시 1명** |
| **C**onsulted | 의견을 듣는 사람 | 0명 이상 |
| **I**nformed | 결과만 통보받는 사람 | 0명 이상 |

> 한 작업에 A가 2명 이상이면 결정이 안 된다. **반드시 1명**이 A다.

RACI를 모든 작업에 매기는 것은 과도하다. **핵심 결정 작업 5~10개**에만 매긴다.

````python
raci = pd.DataFrame([
    ["1.2 Online application form",  "IT Lead",   "Director", "Comm Lead, Ops Lead", "All staff"],
    ["1.3 Document review",          "Ops Lead",  "Director", "Legal",               "Comm Lead"],
    ["2.2 Run cohort",               "Edu Lead",  "Director", "External trainers",   "All staff"],
    ["3.2 Interview matching",       "Biz Lead",  "Director", "Edu Lead",            "All staff"],
    ["4.2 12-month outcome report",  "Eval Lead", "Director", "Ops Lead",            "Board"],
], columns=["task", "Responsible", "Accountable", "Consulted", "Informed"])
raci
````

## 5. 리스크 식별과 매트릭스

**식별 매뉴얼 (5단계)**

1. WBS 각 흐름마다 "무엇이 잘못될 수 있는가" 브레인스토밍.
2. 분류: 기술/일정/비용/자원/외부/조직.
3. 발생 확률 (0~1).
4. 영향도 (1~5).
5. **리스크 점수 = 확률 × 영향**.

**매트릭스**

```
영향도 높음
   |
   |  주의 관찰      ★ 즉시 대응
   |  (Watch)         (Act now)
   |
   +-------------+-------------
   |
   |  무시 가능      완화 검토
   |                  (Mitigate)
   |
   +-------------------- 발생 확률 높음
```

````python
risks = pd.DataFrame([
    ["R1", "Application volume below target",  "Schedule",       0.5, 4, "Mitigate", "Add 2 outreach channels",     "Comm Lead", "Apply < 100 by week 6"],
    ["R2", "Training drop-out",                "Resource",       0.4, 3, "Mitigate", "Stipend + mentor",            "Edu Lead",  "Drop-out > 25%"],
    ["R3", "Employer cancellation",            "External",       0.3, 5, "Transfer", "Backup partners + MOU",       "Biz Lead",  "1+ partner cancels"],
    ["R4", "Budget overrun",                   "Cost",           0.4, 4, "Mitigate", "Quarterly budget gate",       "CFO",       "Spend > +10% YTD"],
    ["R5", "Eligibility regulation change",    "External",       0.2, 5, "Avoid",    "Lobby + flexibility clause", "Director",  "Bill introduced"],
    ["R6", "Data privacy incident",            "Technical",      0.2, 5, "Avoid",    "PIA + access control",        "IT Lead",   "Audit finding"],
    ["R7", "Staff turnover",                   "Organizational", 0.5, 3, "Mitigate", "Backup roles, doc",           "HR",        "1+ leader resigns"],
    ["R8", "Regional economic shock",          "External",       0.2, 4, "Accept",   "Monitor monthly indicators", "Director",  "Local UR > +2pp"],
], columns=["id", "description", "category", "probability", "impact", "strategy", "response", "owner", "trigger"])

risks["score"] = risks["probability"] * risks["impact"]
risks.sort_values("score", ascending=False)
````

````python
plt.figure(figsize=(7, 6))
plt.scatter(
    risks["probability"], risks["impact"],
    s=140, facecolors="white", edgecolors="black", linewidths=1.2,
)
for _, row in risks.iterrows():
    plt.annotate(
        row["id"], (row["probability"], row["impact"]),
        xytext=(5, 5), textcoords="offset points",
        color="black", fontsize=9,
    )
plt.axhline(3, color="gray", linestyle="--", linewidth=0.8)
plt.axvline(0.4, color="gray", linestyle="--", linewidth=0.8)
plt.xlim(0, 1)
plt.ylim(0, 5.5)
plt.xlabel("Probability", color="black")
plt.ylabel("Impact (1-5)", color="black")
plt.title("Risk Matrix", color="black")
plt.tight_layout()
plt.show()
````

## 6. 4대 대응 전략 매뉴얼

| 전략 | 의미 | 예시 |
|------|------|------|
| **회피 (Avoid)** | 리스크 원인 자체 제거 | 위험 지역 사업 철수 |
| **완화 (Mitigate)** | 확률·영향 감소 | 백업, 다중 공급선, 사전 점검 |
| **전가 (Transfer)** | 제3자에게 이전 | 보험, 외주, 계약 조항 |
| **수용 (Accept)** | 그대로 받음 | 작은 위험은 모니터만 |

**선택 매뉴얼 (간단판)**

```
영향 큼 + 확률 높음 → 회피 우선, 안 되면 완화
영향 큼 + 확률 낮음 → 전가 (보험 적합)
영향 작음 + 확률 높음 → 완화
영향 작음 + 확률 낮음 → 수용 (모니터링)
```

**실수 패턴**

- 모든 리스크에 "완화"를 적는다 → 우선순위가 사라진다.
- 영향 큰 리스크에 "수용"을 적는다 → 사고 시 변명 불가.
- 대응안이 추상적이다 ("잘 관리한다") → 실행 불가.

## 7. 조기 경보 지표 — 11·12주차 연결

- R4 → "YTD 예산 +10% 초과"

## 8. 바이브 코딩 프롬프트

**1단계: OKR**

```text
내 사업/정책 주제는 [주제명]이야.
Objective 한 문장과 Key Results 3~5개를 만들어줘.
각 KR은 SMART 형식이어야 해.
```

**2단계: WBS**

```text
위 OKR을 달성하기 위한 WBS를 만들어줘.
Level 2는 4~6개, 각 Level 2 아래에 Level 3 3~4개로 쪼개고
각 Level 3 작업에 예상 기간(주), 담당 역할, 산출물을 적어줘.
```

**3단계: RACI**

```text
위 WBS에서 핵심 결정 작업 5~7개를 골라 RACI 매트릭스를 만들어줘.
Accountable은 반드시 1명만 적어줘.
```

**4단계: 리스크 레지스터**

```text
이 사업에서 발생할 수 있는 리스크 8개를 식별해줘.
각 리스크에 분류, 발생 확률(0-1), 영향(1-5), 점수, 4가지 전략(회피/완화/전가/수용) 중 적절한 것, 구체 대응안, 담당자, 조기 경보 트리거를 적어줘.
표로 정리해줘.
```

**5단계: 매트릭스 시각화**

```text
위 리스크 표를 확률(x)과 영향(y) 산점도로 그리는 Python 코드를 만들어줘.
각 점에 ID 라벨을 붙이고, 4분면을 회색 점선으로 나눠줘. 흑백으로.
```
