## 1. 예측 vs 시나리오 플래닝

예측은 "가장 그럴듯한 미래 하나"를 맞히려고 한다. 격변기에는 빗나간다.

시나리오 플래닝은 "가능한 미래 3-4개"를 그려 두고, 어느 것이 와도 **무너지지 않을 전략**을 찾는다.

| 구분 | 예측 (Forecasting) | 시나리오 플래닝 |
|------|------------------|----------------|
| 목적 | 가장 가능성 높은 미래 1개 | 가능한 미래 3-4개 |
| 가정 | 추세가 이어진다 | 추세가 끊길 수 있다 |
| 결과물 | 숫자 하나 | 서사 여러 개 |
| 적합 상황 | 안정적 환경 | 격변기·구조 변화 |

**기획자의 질문**

> 어떤 미래가 와도 살아남는 전략은 무엇인가? 어떤 미래에서만 통하는 전략은 무엇인가?

## 2. 동인 식별: STEEP 프레임워크

동인(Driving Forces)은 미래를 결정하는 큰 흐름이다. 빠뜨리지 않으려면 5개 영역으로 나눠 본다.

| STEEP | 영역 | 예시 |
|-------|------|------|
| **S**ocial | 사회·인구 | 고령화, 세대 가치관 |
| **T**echnological | 기술 | 자율주행, AI |
| **E**conomic | 경제 | 금리, 환율, 성장률 |
| **E**cological | 환경·기후 | 탄소세, 기후 재해 |
| **P**olitical | 정치·규제 | 선거, 규제 강화 |

**매뉴얼 (8단계)**

1. 주제와 시간 지평을 정한다 (예: "2035년 한국 도시 모빌리티").
2. STEEP 5개 영역마다 동인을 2~3개씩 적는다 (총 10~15개).
3. 각 동인에 **불확실성 점수(0-10)** 와 **영향도 점수(0-10)** 를 매긴다.
4. 표로 정리한다.
5. 영향도 큰 것부터 살핀다.
6. 영향도와 불확실성 모두 높은 동인 중 **서로 독립적인 두 개**를 시나리오 축으로 고른다.
7. 4시나리오를 만든다.
8. 각 시나리오에 서사와 조기 경보 지표를 붙인다.

````python
# 기본 라이브러리 불러오기
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

## 3. 실습 데이터: 2035년 도시 모빌리티 동인 12개

오늘은 수업용 가상 데이터를 사용한다. 실제 분석에서는 AI에게 동인 식별을 의뢰하고 면담·자료로 검증해야 한다.

````python
forces = pd.DataFrame({
    "force": [
        "EV battery cost decline",
        "Charging infrastructure rollout",
        "Carbon regulation tightening",
        "Renewable electricity share",
        "Aging population in cities",
        "Remote work adoption",
        "Autonomous driving readiness",
        "Used car market growth",
        "Lithium supply risk",
        "Public transit investment",
        "Shared mobility platforms",
        "Anti-EV policy backlash",
    ],
    "category": [
        "Tech", "Tech", "Political", "Ecological",
        "Social", "Social", "Tech", "Economic",
        "Economic", "Political", "Social", "Political",
    ],
    "uncertainty": [4, 7, 8, 5, 3, 6, 8, 4, 9, 6, 6, 7],
    "impact":      [8, 9, 9, 7, 5, 5, 8, 4, 6, 7, 6, 7],
})
forces
````

## 4. 불확실성-영향도 매트릭스

두 점수를 두 축으로 두고 산점도로 그린다. 4분면 의미는 다음과 같다.

```
영향도 높음
   |
   |  결정 요인         ★ 핵심 불확실성
   |  (이미 정해짐)        (시나리오 축)
   |
   +-------------+-------------
   |
   |  무시 가능         관심사
   |                   (배경 변수)
   |
   +-------------------- 불확실성 높음
```

오른쪽 위(영향도·불확실성 모두 높음)에 있는 동인이 **시나리오 축 후보**다.

````python
plt.figure(figsize=(8, 6))
plt.scatter(
    forces["uncertainty"], forces["impact"],
    s=130, facecolors="white", edgecolors="black", linewidths=1.2,
)
for _, row in forces.iterrows():
    plt.annotate(
        row["force"], (row["uncertainty"], row["impact"]),
        xytext=(5, 5), textcoords="offset points",
        color="black", fontsize=8,
    )
plt.axhline(5.5, color="gray", linestyle="--", linewidth=0.8)
plt.axvline(5.5, color="gray", linestyle="--", linewidth=0.8)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xlabel("Uncertainty", color="black")
plt.ylabel("Impact", color="black")
plt.title("Uncertainty-Impact Matrix", color="black")
plt.tight_layout()
plt.show()

# 점수 곱이 높은 순으로 시나리오 축 후보 추천
ranked = forces.copy()
ranked["score"] = ranked["uncertainty"] * ranked["impact"]
ranked.sort_values("score", ascending=False).head(5)
````

## 5. 2×2 시나리오 매트릭스

이번 예시에서는 두 축을 다음과 같이 정한다.

- **X축**: Carbon regulation tightening — 강화 ↔ 약화
- **Y축**: Charging infrastructure rollout — 빠름 ↔ 느림

두 축의 두 극단을 곱하면 4개 시나리오가 나온다.

**4개 시나리오 (예시)**

| 사분면 | 이름 | 핵심 서사 |
|--------|------|----------|
| 규제 ↑ · 충전 빠름 | **Green Sprint** | 정부가 강한 탄소 규제를 도입하고 충전 인프라가 폭발적으로 늘어난다. EV 전환이 가속된다. |
| 규제 ↓ · 충전 빠름 | **Tech Race** | 정부 규제는 약하지만 민간 기술 경쟁으로 충전망이 빠르게 깔린다. 시장 주도형 전환. |
| 규제 ↑ · 충전 느림 | **Stuck in Traffic** | 규제는 강한데 인프라가 따라오지 못해 사용자 불만이 폭발한다. 정치적 백래시. |
| 규제 ↓ · 충전 느림 | **Slow Transition** | 모든 변화가 미뤄진다. 기존 자동차 산업이 연장된다. |

**서사 작성 팁**

- 각 시나리오마다 "누가 이긴다"와 "누가 진다"를 명시한다.
- 시나리오 이름은 짧고 그림이 그려져야 한다.
- 확률을 매기지 않는다. 가능성을 펼쳐 두는 것이 목적이다.

````python
scenarios = pd.DataFrame({
    "name": ["Green Sprint", "Tech Race", "Stuck in Traffic", "Slow Transition"],
    "carbon_regulation": ["+", "-", "+", "-"],
    "charging_rollout": ["+", "+", "-", "-"],
})

# 4분면 그리기
plt.figure(figsize=(7, 6))
label_pos = {"Green Sprint": (0.75, 0.75), "Tech Race": (0.25, 0.75),
             "Stuck in Traffic": (0.75, 0.25), "Slow Transition": (0.25, 0.25)}
for name, (x, y) in label_pos.items():
    plt.text(x, y, name, ha="center", va="center", color="black", fontsize=12,
             bbox=dict(boxstyle="round,pad=0.5", facecolor="white", edgecolor="black"))
plt.axhline(0.5, color="black", linewidth=1)
plt.axvline(0.5, color="black", linewidth=1)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.text(0.5, -0.08, "Carbon regulation: weak  <-->  strong", ha="center", color="black")
plt.text(-0.08, 0.5, "Charging: slow  <-->  fast", va="center", rotation=90, color="black")
plt.xticks([])
plt.yticks([])
plt.title("2x2 Scenario Matrix", color="black")
plt.tight_layout()
plt.show()

scenarios
````

## 6. 윈드 터널링: 강건한 전략 찾기

윈드 터널링은 풍동(風洞) 실험에서 따온 말이다. 여러 시나리오라는 "바람"에 전략을 통과시켜, 어느 바람에도 무너지지 않는 전략을 찾는다.

**매뉴얼**

1. 전략 후보 3~5개를 적는다.
2. 각 전략을 시나리오마다 점수 매긴다 ( -2 ~ +2 ).
3. **최저 점수(min)** 와 **평균 점수(mean)** 두 가지를 본다.
4. min이 가장 높은 전략이 가장 강건하다.
5. mean이 높지만 min이 낮은 전략은 "잘 될 때는 크지만 무너질 위험도 큰" 도박형이다.

````python
wind_tunnel = pd.DataFrame(
    [
        ["A: Aggressive EV factory",   2,  -2,  1, -1],
        ["B: Hybrid + transit invest", 1,   1,  1,  2],
        ["C: Wait & license tech",    -1,   2, -1,  1],
    ],
    columns=["strategy", "Green Sprint", "Tech Race", "Stuck in Traffic", "Slow Transition"],
)

wind_tunnel["min"] = wind_tunnel.iloc[:, 1:].min(axis=1)
wind_tunnel["mean"] = wind_tunnel.iloc[:, 1:5].mean(axis=1)
wind_tunnel.sort_values("min", ascending=False)
````

**해석 예시**

- 전략 B (`min=1, mean=1.25`): 어느 시나리오에서도 무너지지 않는다. **강건한 전략**.
- 전략 A (`min=-2`): Tech Race에서는 상위지만 Stuck in Traffic에서는 큰 손실. **도박형**.
- 전략 C (`min=-1`): 평균은 낮고 두 시나리오에서 음수. **약한 전략**.

## 7. 조기 경보 지표

각 시나리오가 현실이 되어 가는지 미리 알려주는 신호다.

**지표 설계 매뉴얼**

1. 시나리오마다 "이게 보이면 그 시나리오가 진행 중"이라고 할 만한 지표를 2~3개 적는다.
2. 측정 가능한 형태로 쓴다 ("EV 등록 비율", "충전소 증가율" 등).
3. 데이터 출처를 적는다 (어디서 분기마다 볼 수 있는가).
4. 임계값을 정한다 ("분기 등록 비율 30% 초과 시").

````python
early_warning = pd.DataFrame([
    ["Green Sprint",     "EV market share > 30% by 2028",         "Quarterly registrations"],
    ["Green Sprint",     "Charging stations growth > 25% per yr", "City data dashboard"],
    ["Tech Race",        "OEM autonomous pilots in 5+ cities",    "Industry news scan"],
    ["Tech Race",        "Private charging investments > $X bn",  "Industry filings"],
    ["Stuck in Traffic", "Charging stations growth < 5% per yr",  "City data dashboard"],
    ["Stuck in Traffic", "Public complaints rising",              "Citizen petition tracker"],
    ["Slow Transition",  "Carbon tax delayed past 2027",          "Government bill tracker"],
    ["Slow Transition",  "ICE vehicle sales flat or rising",      "KAIDA monthly reports"],
], columns=["scenario", "indicator", "data_source"])

early_warning
````

**2단계: 두 축 추천**


