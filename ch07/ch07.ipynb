## 1. 이해관계자란?

이해관계자는 정책이나 사업의 결과에 영향을 받거나 결과에 영향을 줄 수 있는 사람·집단이다.

예를 들어 청년 주거 정책에는 다음과 같은 이해관계자가 있다.

- 시장과 주거 담당 부서
- 청년 단체와 청년 본인
- 임대인, 건설사, 부동산 중개
- 지역 주민과 시민 단체
- 언론과 전문가

핵심 질문은 하나다.

> 누가 영향을 받는가, 그리고 누가 결과를 바꿀 힘을 가지고 있는가?

## 2. 영향력-관심도 매트릭스

이해관계자를 두 가지 축으로 평가한다.

- **영향력 (Power)**: 결정을 바꾸거나 막을 수 있는 힘
- **관심도 (Interest)**: 이 정책에 얼마나 관심이 있는가

두 축으로 4분면을 만든다.

```
영향력 높음
   |
   |  관리              ★ 핵심 관계자
   |  (Manage)            (Key Player)
   |
   +------------+-----------
   |
   |  최소 관심         정보 제공
   |  (Monitor)         (Keep Informed)
   |
   +-------------------- 관심도 높음
```

기획자는 4분면에 따라 대응 방식을 다르게 한다. 핵심 관계자에게는 시간을 많이 쓰고, 최소 관심 그룹은 변화만 살핀다.

## 3. 정책 네트워크의 직관

이해관계자는 따로 떨어져 있지 않다. 누가 누구에게 영향을 주는지 화살표로 이으면 네트워크가 된다.

쉽게 말해 다음과 같다.

- 화살표가 많이 들어오는 사람: 다른 사람의 의견을 많이 받는 위치
- 화살표가 많이 나가는 사람: 다른 사람에게 영향을 주는 위치
- 중심에 있는 사람: 작은 행동에도 파급력이 크다

기획자는 중심에 있는 사람부터 설득한다.

````python
# 기본 라이브러리 불러오기
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
````

## 5. 실습 데이터

실제 정책 사례에서 이해관계자 데이터를 수집하면 면담과 자료 조사에 시간이 많이 든다. 오늘은 수업용 가상 데이터를 사용한다.

주제는 `청년 주거 지원 정책`이다. 영향력과 관심도는 0-10 점수다.

````python
stakeholders = pd.DataFrame({
    "name": [
        "City Mayor", "Housing Dept", "Youth Council", "Local Press",
        "Construction Co.", "Resident Group", "University Lab", "Bank",
    ],
    "power":    [9, 8, 4, 6, 7, 5, 3, 6],
    "interest": [7, 9, 9, 6, 5, 8, 6, 3],
})

stakeholders
````

## 6. 영향력-관심도 매트릭스 그리기

두 점수를 두 축으로 두고 산점도로 그린다. 5.5를 기준선으로 4분면을 나눈다.

이 코드는 AI에게 그대로 요청해도 된다. 중요한 것은 코드 암기가 아니라 결과 해석이다.

````python
def label_quadrant(row):
    if row["power"] >= 6 and row["interest"] >= 6:
        return "Key Player"
    if row["power"] >= 6:
        return "Manage"
    if row["interest"] >= 6:
        return "Keep Informed"
    return "Monitor"

stakeholders["quadrant"] = stakeholders.apply(label_quadrant, axis=1)

plt.figure(figsize=(7, 6))
plt.scatter(
    stakeholders["interest"], stakeholders["power"],
    s=120, facecolors="white", edgecolors="black", linewidths=1.2,
)
for _, row in stakeholders.iterrows():
    plt.annotate(
        row["name"], (row["interest"], row["power"]),
        xytext=(6, 6), textcoords="offset points",
        color="black", fontsize=9,
    )

plt.axhline(5.5, color="gray", linestyle="--", linewidth=0.8)
plt.axvline(5.5, color="gray", linestyle="--", linewidth=0.8)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xlabel("Interest", color="black")
plt.ylabel("Power", color="black")
plt.title("Power-Interest Matrix", color="black")
plt.tight_layout()
plt.show()

stakeholders.sort_values(["quadrant", "power"], ascending=[True, False])
````

## 7. 누가 누구에게 영향을 주는가

이번에는 이해관계자 사이의 영향 관계를 표로 정리한다. 화살표는 `A → B`로 읽는다 ("A가 B에게 영향을 준다").

그리고 가장 많이 연결된 사람을 찾는다. 이를 `중심성(centrality)`이라고 부른다. 가장 단순한 지표는 연결된 선의 수, 즉 `degree centrality`다.

````python
influence_edges = [
    ("City Mayor", "Housing Dept"),
    ("Housing Dept", "Construction Co."),
    ("Housing Dept", "Resident Group"),
    ("Youth Council", "City Mayor"),
    ("Local Press", "City Mayor"),
    ("Resident Group", "Local Press"),
    ("University Lab", "Housing Dept"),
    ("Bank", "Construction Co."),
]

G = nx.DiGraph()
G.add_edges_from(influence_edges)

pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(7, 6))
nx.draw_networkx_nodes(G, pos, node_color="white", edgecolors="black", linewidths=1.2, node_size=900)
nx.draw_networkx_edges(G, pos, edge_color="gray", arrows=True)
nx.draw_networkx_labels(G, pos, font_color="black", font_size=8)
plt.title("Stakeholder Influence Network", color="black")
plt.axis("off")
plt.tight_layout()
plt.show()

degree = nx.degree_centrality(G)
centrality = (
    pd.DataFrame({"name": list(degree.keys()), "degree_centrality": list(degree.values())})
    .sort_values("degree_centrality", ascending=False)
    .reset_index(drop=True)
)
centrality
````

## 8. 기획 관점으로 해석하기

숫자만 보고 끝내면 분석이다. 기획은 다음 행동까지 생각해야 한다.

다음 표는 4분면을 `대응 전략`과 `다음 행동`으로 바꾸는 예시다.

````python
actions = {
    "Key Player":    "Hold weekly meetings and co-design the plan.",
    "Manage":        "Brief regularly so they do not block decisions.",
    "Keep Informed": "Share updates so they support the policy publicly.",
    "Monitor":       "Watch only; act if their position changes.",
}

planning_table = stakeholders.copy()
planning_table["next_action"] = planning_table["quadrant"].map(actions)
planning_table[["name", "power", "interest", "quadrant", "next_action"]]
````
