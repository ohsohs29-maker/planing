## 1. 이해관계자란?

이해관계자는 정책이나 사업의 결과에 영향을 받거나 결과에 영향을 줄 수 있는 사람·집단이다.

예를 들어 청년 주거 정책에는 다음과 같은 이해관계자가 있다.

- 시장과 주거 담당 부서
- 청년 단체와 청년 본인
- 임대인, 건설사, 부동산 중개
- 지역 주민과 시민 단체
- 언론과 전문가

**기획자의 질문**

> 누가 영향을 받는가, 그리고 누가 결과를 바꿀 힘을 가지고 있는가?

이 두 질문을 분리하는 것이 분석의 출발점이다.

## 2. 이해관계자 유형 분류 매뉴얼

빠짐없이 식별하기 위해 4가지 역할 축으로 점검한다.

| 역할 축 | 질문 | 예시 |
|---------|-----|------|
| **결정자 (Decider)** | 누가 최종 승인 권한을 가지나 | 시장, 의회, 이사회 |
| **실행자 (Executor)** | 누가 실제로 정책을 운영하나 | 담당 부서, 위탁 운영사 |
| **수혜·피해자 (Affected)** | 누가 결과를 직접 받나 | 청년, 입주자, 임대인 |
| **여론 형성자 (Influencer)** | 누가 다른 사람의 인식을 바꾸나 | 언론, 시민 단체, 인플루언서 |

**작성 매뉴얼**

1. 4축에 각각 2~3명씩 적는다 (총 8~12명).
2. 빠진 축이 있으면 "왜 비었지?" 물어본다 — 빠진 축이 사고의 사각이 된다.
3. 각 인물·집단에 영향력(0-10), 관심도(0-10) 점수를 매긴다.
4. 다음 단계: 매트릭스 → 네트워크 → 전략.

## 3. 영향력-관심도 매트릭스

이해관계자를 두 축으로 평가한다.

- **영향력 (Power)**: 결정을 바꾸거나 막을 수 있는 힘
- **관심도 (Interest)**: 이 정책에 얼마나 관심이 있는가

두 축으로 4분면을 만든다.

```
영향력 높음
   |
   |  관리              ★ 핵심 관계자
   |  (Manage)            (Key Player)
   |
   +------------+-----------
   |
   |  최소 관심         정보 제공
   |  (Monitor)         (Keep Informed)
   |
   +-------------------- 관심도 높음
```

**4분면별 운용 매뉴얼**

| 분면 | 관리 강도 | 주된 활동 | 위험 신호 |
|------|---------|---------|----------|
| **Key Player** (영향↑·관심↑) | 매우 높음 | 주간 미팅, 공동 설계, 쟁점 사전 조율 | 회의에 안 나오기 시작 |
| **Manage** (영향↑·관심↓) | 높음 | 분기 브리핑, 핵심 결정 사전 합의 | 갑작스런 반대 |
| **Keep Informed** (영향↓·관심↑) | 중간 | 뉴스레터·공청회·SNS 업데이트 | 부정적 여론 형성 |
| **Monitor** (영향↓·관심↓) | 낮음 | 동향만 살핌 | 다른 분면으로 이동 |

## 4. 정책 네트워크의 직관

이해관계자는 따로 떨어져 있지 않다. **누가 누구에게 영향을 주는지**를 화살표로 이으면 네트워크가 된다.

쉽게 말해 다음과 같다.

- 화살표가 많이 들어오는 사람: 다른 사람의 의견을 많이 받는 위치 — "수신자"
- 화살표가 많이 나가는 사람: 다른 사람에게 영향을 주는 위치 — "발신자"
- 양쪽 다 많은 사람: 정보·영향의 허브 — "중계자"

기획자는 중심에 있는 사람을 우선 확인하되, 실제 권한·입장·이해관계를 함께 봐야 한다. 네트워크 중심성은 설득 순서를 자동으로 정해 주는 답이 아니라, 누구를 먼저 살펴볼지 알려 주는 단서다.

## 5. 중심성 3가지 직관 비교

수학을 외울 필요 없다. 이 표만 보고 고르면 된다.

| 중심성 | 무엇을 잡나 | 언제 쓰나 |
|--------|-----------|---------|
| **Degree** (전체 연결 수) | 가장 많은 사람과 연결된 사람 | 일반적인 "중심에 누가 있나" 물음 |
| **In-degree** (들어오는 화살표 수) | 다른 사람의 의견·압력을 많이 받는 사람 | 결정자·수신자 후보 탐색 |
| **Out-degree** (나가는 화살표 수) | 다른 사람에게 많이 영향을 주는 사람 | 영향 확산자 후보 탐색 |

**한 줄 요약**

> Degree는 "누가 많이 연결되어 있나", In은 "누가 의견·압력을 많이 받나", Out은 "누가 영향을 많이 보내나"를 보는 단서다.

## 오늘의 핵심

이해관계자 분석은 모든 사람을 똑같이 대하는 일이 아니다. **누구에게 시간을 더 쓸지** 정하는 일이고, **누가 다른 사람의 입장을 바꿀 수 있는지** 찾는 일이다. 바이브 코딩은 이 판단을 대신해 주는 것이 아니라, 표와 그래프를 빨리 만들고 사람이 검토할 시간을 확보하는 방법이다.
