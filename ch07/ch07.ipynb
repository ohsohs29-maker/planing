## 1. 이해관계자란?

이해관계자는 정책이나 사업의 결과에 영향을 받거나 결과에 영향을 줄 수 있는 사람·집단이다.

예를 들어 청년 주거 정책에는 다음과 같은 이해관계자가 있다.

- 시장과 주거 담당 부서
- 청년 단체와 청년 본인
- 임대인, 건설사, 부동산 중개
- 지역 주민과 시민 단체
- 언론과 전문가

**기획자의 질문**

> 누가 영향을 받는가, 그리고 누가 결과를 바꿀 힘을 가지고 있는가?

이 두 질문을 분리하는 것이 분석의 출발점이다.

## 2. 이해관계자 유형 분류 매뉴얼

빠짐없이 식별하기 위해 4가지 역할 축으로 점검한다.

| 역할 축 | 질문 | 예시 |
|---------|-----|------|
| **결정자 (Decider)** | 누가 최종 승인 권한을 가지나 | 시장, 의회, 이사회 |
| **실행자 (Executor)** | 누가 실제로 정책을 운영하나 | 담당 부서, 위탁 운영사 |
| **수혜·피해자 (Affected)** | 누가 결과를 직접 받나 | 청년, 입주자, 임대인 |
| **여론 형성자 (Influencer)** | 누가 다른 사람의 인식을 바꾸나 | 언론, 시민 단체, 인플루언서 |

**작성 매뉴얼**

1. 4축에 각각 2~3명씩 적는다 (총 8~12명).
2. 빠진 축이 있으면 "왜 비었지?" 물어본다 — 빠진 축이 사고의 사각이 된다.
3. 각 인물·집단에 영향력(0-10), 관심도(0-10) 점수를 매긴다.
4. 다음 단계: 매트릭스 → 네트워크 → 전략.

## 3. 영향력-관심도 매트릭스

이해관계자를 두 축으로 평가한다.

- **영향력 (Power)**: 결정을 바꾸거나 막을 수 있는 힘
- **관심도 (Interest)**: 이 정책에 얼마나 관심이 있는가

두 축으로 4분면을 만든다.

```
영향력 높음
   |
   |  관리              ★ 핵심 관계자
   |  (Manage)            (Key Player)
   |
   +------------+-----------
   |
   |  최소 관심         정보 제공
   |  (Monitor)         (Keep Informed)
   |
   +-------------------- 관심도 높음
```

**4분면별 운용 매뉴얼**

| 분면 | 시간 배분 | 주된 활동 | 위험 신호 |
|------|---------|---------|----------|
| **Key Player** (영향↑·관심↑) | 가장 많이 | 주간 미팅, 공동 설계, 결정 권한 일부 공유 | 회의에 안 나오기 시작 |
| **Manage** (영향↑·관심↓) | 정기적 | 분기 브리핑, 핵심 결정 사전 합의 | 갑작스런 반대 |
| **Keep Informed** (영향↓·관심↑) | 주기적 | 뉴스레터·공청회·SNS 업데이트 | 부정적 여론 형성 |
| **Monitor** (영향↓·관심↓) | 최소 | 동향만 살핌 | 다른 분면으로 이동 |

## 4. 정책 네트워크의 직관

이해관계자는 따로 떨어져 있지 않다. **누가 누구에게 영향을 주는지**를 화살표로 이으면 네트워크가 된다.

쉽게 말해 다음과 같다.

- 화살표가 많이 들어오는 사람: 다른 사람의 의견을 많이 받는 위치 — "수신자"
- 화살표가 많이 나가는 사람: 다른 사람에게 영향을 주는 위치 — "발신자"
- 양쪽 다 많은 사람: 정보·영향의 허브 — "중계자"

기획자는 중심에 있는 사람부터 설득한다. 가장자리에 있는 사람을 직접 설득하려 들면 시간만 든다.

## 5. 중심성 3가지 직관 비교

수학을 외울 필요 없다. 이 표만 보고 고르면 된다.

| 중심성 | 무엇을 잡나 | 언제 쓰나 |
|--------|-----------|---------|
| **Degree** (전체 연결 수) | 가장 많은 사람과 연결된 사람 | 일반적인 "중심에 누가 있나" 물음 |
| **In-degree** (들어오는 화살표 수) | 다른 사람의 의견을 가장 많이 받는 사람 | 결정자 식별 |
| **Out-degree** (나가는 화살표 수) | 가장 많이 영향을 주는 사람 | 여론 주도자 식별 |

**한 줄 요약**

> Degree는 "누가 잘 알려져 있나", In은 "누가 결정하나", Out은 "누가 분위기를 만드나".

```text
각자에게 영향력(0-10)과 관심도(0-10) 점수를 매겨주고, 그 점수를 매긴 근거를 한 줄로 적어줘.
```


````python
# 기본 라이브러리 불러오기
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
````

## 7. 실습 데이터: 청년 주거 지원 정책

수업용 가상 데이터다. 영향력과 관심도는 0-10 점수.

역할 축까지 함께 적어 두면 빠진 사람을 찾기 쉽다.

````python
stakeholders = pd.DataFrame({
    "name": [
        "City Mayor", "Housing Dept", "Youth Council", "Local Press",
        "Construction Co.", "Resident Group", "University Lab", "Bank",
    ],
    "role": [
        "Decider", "Executor", "Affected", "Influencer",
        "Executor", "Affected", "Influencer", "Executor",
    ],
    "power":    [9, 8, 4, 6, 7, 5, 3, 6],
    "interest": [7, 9, 9, 6, 5, 8, 6, 3],
})
stakeholders
````

## 8. 영향력-관심도 매트릭스 그리기

두 점수를 두 축으로 두고 산점도로 그린다. 5.5를 기준선으로 4분면을 나눈다.

````python
def label_quadrant(row):
    if row["power"] >= 6 and row["interest"] >= 6:
        return "Key Player"
    if row["power"] >= 6:
        return "Manage"
    if row["interest"] >= 6:
        return "Keep Informed"
    return "Monitor"

stakeholders["quadrant"] = stakeholders.apply(label_quadrant, axis=1)

plt.figure(figsize=(7, 6))
plt.scatter(
    stakeholders["interest"], stakeholders["power"],
    s=120, facecolors="white", edgecolors="black", linewidths=1.2,
)
for _, row in stakeholders.iterrows():
    plt.annotate(
        row["name"], (row["interest"], row["power"]),
        xytext=(6, 6), textcoords="offset points",
        color="black", fontsize=9,
    )
plt.axhline(5.5, color="gray", linestyle="--", linewidth=0.8)
plt.axvline(5.5, color="gray", linestyle="--", linewidth=0.8)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xlabel("Interest", color="black")
plt.ylabel("Power", color="black")
plt.title("Power-Interest Matrix", color="black")
plt.tight_layout()
plt.show()

stakeholders.sort_values(["quadrant", "power"], ascending=[True, False])
````

## 9. 분면별 운용 전략 표

분면이 정해지면 다음 행동이 거의 자동으로 결정된다. 분면별 표준 행동을 매뉴얼처럼 적어 둔다.

````python
actions = {
    "Key Player":    "Weekly meetings, co-design the plan, share decision rights",
    "Manage":        "Quarterly briefings, secure buy-in before announcing",
    "Keep Informed": "Monthly newsletters, public hearings, social media updates",
    "Monitor":       "Watch only; reassign if their position changes",
}
time_share = {
    "Key Player":    "35%",
    "Manage":        "30%",
    "Keep Informed": "25%",
    "Monitor":       "10%",
}

planning_table = stakeholders.copy()
planning_table["time_share"] = planning_table["quadrant"].map(time_share)
planning_table["next_action"] = planning_table["quadrant"].map(actions)
planning_table[["name", "role", "power", "interest", "quadrant", "time_share", "next_action"]]
````

## 10. 누가 누구에게 영향을 주는가

이번에는 이해관계자 사이의 영향 관계를 표로 정리한다. 화살표는 `A → B`로 읽는다 ("A가 B에게 영향을 준다").

관계를 모를 때는 다음 질문으로 시작한다.

1. 누가 누구의 "위"에 있는가 (조직 위계)?
2. 누가 누구에게 자원·정보를 주는가?
3. 누가 누구의 평판을 바꿀 수 있는가?

````python
influence_edges = [
    ("City Mayor", "Housing Dept"),
    ("Housing Dept", "Construction Co."),
    ("Housing Dept", "Resident Group"),
    ("Youth Council", "City Mayor"),
    ("Local Press", "City Mayor"),
    ("Resident Group", "Local Press"),
    ("University Lab", "Housing Dept"),
    ("Bank", "Construction Co."),
]

G = nx.DiGraph()
G.add_edges_from(influence_edges)

pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(7, 6))
nx.draw_networkx_nodes(G, pos, node_color="white", edgecolors="black", linewidths=1.2, node_size=900)
nx.draw_networkx_edges(G, pos, edge_color="gray", arrows=True)
nx.draw_networkx_labels(G, pos, font_color="black", font_size=8)
plt.title("Stakeholder Influence Network", color="black")
plt.axis("off")
plt.tight_layout()
plt.show()
````

## 11. 중심성 3가지 비교

Degree·In-degree·Out-degree를 한 표로 본다. 1위가 같은 사람이면 "명확한 중심"이고, 다르면 "역할 분리"가 일어나는 네트워크다.

````python
degree = nx.degree_centrality(G)
in_deg = nx.in_degree_centrality(G)
out_deg = nx.out_degree_centrality(G)

centrality = pd.DataFrame({
    "name": list(degree.keys()),
    "degree":   [round(degree[n], 3) for n in degree],
    "in_degree": [round(in_deg[n], 3) for n in degree],
    "out_degree": [round(out_deg[n], 3) for n in degree],
}).sort_values("degree", ascending=False).reset_index(drop=True)
centrality
````

````python
# 1위 비교 — 누가 어떤 면에서 중심인가
leaders = {
    "Most connected (degree)":     centrality.sort_values("degree",     ascending=False).iloc[0]["name"],
    "Most listened to (in-degree)":  centrality.sort_values("in_degree",  ascending=False).iloc[0]["name"],
    "Most influential (out-degree)": centrality.sort_values("out_degree", ascending=False).iloc[0]["name"],
}
for k, v in leaders.items():
    print(f"{k}: {v}")
````
