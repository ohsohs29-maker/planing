# 7주차. 시나리오 플래닝

## 오늘의 목표

이번 시간에는 미래를 정확히 맞히는 법을 배우지 않는다. **여러 가능한 미래에 대비하는 사고방식**을 익힌다.

수업이 끝나면 다음을 할 수 있어야 한다.

1. 예측과 시나리오 플래닝의 차이를 한 문장으로 설명한다.
2. STEEP 동인을 식별하고 불확실성-영향도로 우선순위를 매긴다.
3. 2×2 매트릭스로 4개 시나리오를 만든다.
4. 윈드 터널링으로 강건한 전략을 찾는다.
5. 시나리오마다 조기 경보 지표를 설계한다.

## 1. 예측 vs 시나리오 플래닝

예측은 "가장 그럴듯한 미래 하나"를 맞히려고 한다. 격변기에는 빗나간다.

시나리오 플래닝은 "가능한 미래 3-4개"를 그려 두고, 어느 것이 와도 **무너지지 않을 전략**을 찾는다.

| 구분 | 예측 (Forecasting) | 시나리오 플래닝 |
|------|------------------|----------------|
| 목적 | 가장 가능성 높은 미래 1개 | 가능한 미래 3-4개 |
| 가정 | 추세가 이어진다 | 추세가 끊길 수 있다 |
| 결과물 | 숫자 하나 | 서사 여러 개 |
| 적합 상황 | 안정적 환경 | 격변기·구조 변화 |

**기획자의 질문**

> 어떤 미래가 와도 살아남는 전략은 무엇인가? 어떤 미래에서만 통하는 전략은 무엇인가?

## 2. STEEP 프레임워크와 시나리오 플래닝 프로세스

동인(Driving Forces)은 미래를 결정하는 큰 흐름이다. 빠뜨리지 않으려면 5개 영역으로 나눠 본다.

| STEEP | 영역 | 예시 |
|-------|------|------|
| **S**ocial | 사회·인구 | 고령화, 세대 가치관 |
| **T**echnological | 기술 | 자율주행, AI |
| **E**conomic | 경제 | 금리, 환율, 성장률 |
| **E**cological | 환경·기후 | 탄소세, 기후 재해 |
| **P**olitical | 정치·규제 | 선거, 규제 강화 |

**시나리오 플래닝 4단계**

1. STEEP 동인을 식별하고 불확실성·영향도를 매긴다 → 핵심 불확실성 2개를 고른다.
2. 두 축의 조합으로 시나리오 4개를 만들고 서사를 붙인다.
3. 전략 후보를 시나리오에 통과시켜(윈드 터널링) 강건한 전략을 찾는다.
4. 시나리오마다 조기 경보 지표를 설계한다.

## 3. 사례: 2035년 한국 도시 모빌리티

**배경**

서울시 교통정책과에서 2035년 도시 모빌리티 시나리오 플래닝을 의뢰했다. 전기차, 자율주행, 탄소 규제 등 불확실한 변수가 많아 하나의 미래를 예측하기보다 여러 시나리오에 대비하려 한다.

이 사례에서 AI와 대화하며 시나리오 플래닝을 완성하는 과정을 보여준다.

### 1단계. 동인 분석 + 핵심 불확실성 선정

동인마다 **불확실성**(얼마나 결과를 모르는가)과 **영향도**(결과가 얼마나 큰가)를 점수로 매긴다. 둘 다 높은 동인이 **시나리오 축 후보**다.

```
영향도 높음
   |
   |  결정 요인         ★ 핵심 불확실성
   |  (이미 정해짐)        (시나리오 축)
   |
   +-------------+-------------
   |
   |  무시 가능         관심사
   |                   (배경 변수)
   |
   +-------------------- 불확실성 높음
```

**AI에게 이렇게 질문합니다**

> 나는 2035년 한국 도시 모빌리티에 대한 시나리오 플래닝을 하려고 해. STEEP 프레임워크로 동인 12개를 만들고, 각 동인에 불확실성·영향도 점수(0-10)를 매겨줘. score = 불확실성 × 영향도 순으로 정렬하고, 시나리오 축으로 적합한 독립적인 두 동인을 추천해줘.

**AI의 답변**

| 순위 | 동인 | 범주 | 불확실성 | 영향도 | score |
|:---:|------|:---:|:---:|:---:|:---:|
| 1 | 탄소 규제 강화 | 정치 | 8 | 9 | 72 |
| 2 | 자율주행 기술 성숙도 | 기술 | 8 | 8 | 64 |
| 3 | 충전 인프라 확충 속도 | 기술 | 7 | 9 | 63 |
| 4 | 리튬 공급 리스크 | 경제 | 9 | 6 | 54 |
| 5 | 반(反)EV 정책 반발 | 정치 | 7 | 7 | 49 |
| 6 | 대중교통 투자 | 정치 | 6 | 7 | 42 |
| 7 | 공유 모빌리티 플랫폼 | 사회 | 6 | 6 | 36 |
| 8 | 재생에너지 비중 확대 | 환경 | 5 | 7 | 35 |
| 9 | EV 배터리 가격 하락 | 기술 | 4 | 8 | 32 |
| 10 | 원격근무 확산 | 사회 | 6 | 5 | 30 |
| 11 | 중고차 시장 성장 | 경제 | 4 | 4 | 16 |
| 12 | 도시 고령화 | 사회 | 3 | 5 | 15 |

시나리오 축 추천:
- **X축: 탄소 규제 강화** (정치, score 72) — 강화 ↔ 약화
- **Y축: 충전 인프라 확충 속도** (기술, score 63) — 빠름 ↔ 느림

자율주행(score 64)도 후보지만 충전 인프라와 같은 기술 범주여서 상관이 있다. 탄소 규제(정치)와 충전 인프라(기술)는 서로 다른 영역이므로 독립적이다.

**핵심**: 시나리오 축은 score만 높다고 되는 것이 아니다. **서로 독립적인** 두 동인을 골라야 4개 시나리오가 의미 있게 나뉜다.

### 2단계. 2×2 시나리오 매트릭스

두 축의 극단을 조합하면 4개의 미래가 나온다. 각 시나리오에 **짧은 이름**과 **서사**(누가 이기고, 누가 지는가)를 붙인다. 확률은 매기지 않는다 — 가능성을 펼쳐 두는 것이 목적이다.

**AI에게 이렇게 질문합니다**

> 탄소 규제(강화/약화)와 충전 인프라(빠름/느림) 두 축으로 시나리오 4개를 만들어줘. 각 시나리오에 한국어 이름(2-3단어)과 핵심 서사(2문장)를 붙여줘. 서사에는 누가 이기고 누가 지는지 적어줘.

**AI의 답변**

| 시나리오 | 규제 | 인프라 | 핵심 서사 |
|---------|:---:|:---:|----------|
| **녹색 질주** | 강화 ↑ | 빠름 ↑ | 강한 탄소세와 빠른 충전망 확충이 EV 전환을 가속한다. EV 제조사·충전 사업자가 이기고, 내연기관 부품사가 진다. |
| **기술 경쟁** | 약화 ↓ | 빠름 ↑ | 규제는 약하지만 민간 기술 경쟁으로 충전망이 깔린다. 빅테크·스타트업이 이기고, 규제에 기댄 공기업이 진다. |
| **규제 공회전** | 강화 ↑ | 느림 ↓ | 규제는 강한데 인프라가 못 따라와 시민 불만이 폭발한다. 정치적 백래시가 생기고, 정부 신뢰가 진다. |
| **느린 전환** | 약화 ↓ | 느림 ↓ | 모든 변화가 미뤄진다. 기존 자동차 산업·정유사가 시간을 벌고, EV 스타트업이 진다. |

**핵심**: 4개 시나리오는 "좋은 미래 vs 나쁜 미래"가 아니다. **이해관계자마다 유리한 시나리오가 다르다**는 점이 중요하다.

### 3단계. 윈드 터널링: 강건한 전략 찾기

윈드 터널링은 풍동(風洞) 실험에서 따온 말이다. 전략 후보를 여러 시나리오라는 "바람"에 통과시켜, 어느 바람에도 무너지지 않는 전략을 찾는다.

- **min**(최저 점수)이 가장 높은 전략 = **가장 강건한 전략**
- mean은 높지만 min이 낮은 전략 = **도박형 전략**

**AI에게 이렇게 질문합니다**

> 아래 전략 3개를 위 4개 시나리오에서 평가해줘. A: 공격적 EV 공장 투자, B: 하이브리드 + 대중교통 투자, C: 관망 + 기술 라이선스. 각 전략의 시나리오별 점수(-2 ~ +2)를 매기고, min과 mean으로 강건성을 비교해줘.

**AI의 답변**

| 전략 | 녹색 질주 | 기술 경쟁 | 규제 공회전 | 느린 전환 | min | mean |
|------|:---:|:---:|:---:|:---:|:---:|:---:|
| A: 공격적 EV 투자 | +2 | -2 | +1 | -1 | **-2** | 0.00 |
| B: 하이브리드+대중교통 | +1 | +1 | +1 | +2 | **+1** | 1.25 |
| C: 관망+라이선스 | -1 | +2 | -1 | +1 | **-1** | 0.25 |

**추천: 전략 B**

- 전략 B는 min=+1로 어느 시나리오에서도 무너지지 않는다. **강건한 전략**.
- 전략 A는 녹색 질주에서 최고점(+2)이지만 기술 경쟁에서 -2. **도박형**.
- 전략 C는 기술 경쟁에서만 +2, 나머지에서 약하다. **편향된 전략**.

**핵심**: 기획자가 할 일은 "가장 점수가 높은 전략"이 아니라 **"최악의 경우에도 견딜 수 있는 전략"**을 고르는 것이다.

### 4단계. 조기 경보 지표

강건한 전략을 골랐지만, 세상은 바뀐다. 어떤 시나리오가 현실로 다가오는지 알려주는 **신호**를 미리 정해두면, 전략을 적시에 조정할 수 있다.

좋은 지표의 조건:
1. **측정 가능**하다 ("EV 등록 비율" ○, "분위기가 바뀐다" ✕)
2. **데이터 출처**가 있다 (어디서 분기마다 볼 수 있는가)
3. **임계값**이 있다 ("분기 등록 비율 30% 초과 시")

**AI에게 이렇게 질문합니다**

> 4개 시나리오(녹색 질주, 기술 경쟁, 규제 공회전, 느린 전환)가 현실이 되어가는지 알려줄 조기 경보 지표를 시나리오마다 2개씩 만들어줘. 각 지표에 데이터 출처와 임계값을 넣어줘.

**AI의 답변**

| 시나리오 | 지표 | 데이터 출처 | 임계값 |
|---------|------|-----------|--------|
| 녹색 질주 | 신규 EV 등록 비율 | 국토교통부 자동차등록 통계 | 분기 30% 초과 |
| 녹색 질주 | 공공 충전소 증가율 | 한국전력 충전인프라 현황 | 연 25% 초과 |
| 기술 경쟁 | 민간 충전 투자액 | 전자공시시스템(DART) | 연 1조원 초과 |
| 기술 경쟁 | 자율주행 시범 도시 수 | 국토부 자율주행 현황 | 5개 도시 이상 |
| 규제 공회전 | 충전소 증가율 | 한국전력 충전인프라 현황 | 연 5% 미만 |
| 규제 공회전 | 교통 민원 건수 | 국민신문고 | 전년 대비 20% 증가 |
| 느린 전환 | 탄소세 법안 진행 | 국회 의안정보시스템 | 2027년까지 미통과 |
| 느린 전환 | 내연기관 신차 판매량 | 한국자동차산업협회(KAMA) | 전년 대비 횡보·증가 |

**핵심**: 조기 경보 지표는 "미래를 맞히는 도구"가 아니라 **"어떤 시나리오로 가고 있는지 빨리 감지하는 레이더"**다. 분기마다 점검하면 전략 조정 타이밍을 놓치지 않는다.

## 오늘의 핵심

시나리오 플래닝은 미래를 맞히는 일이 아니다. **여러 미래 어디서든 살아남는 전략**을 찾고, **시나리오가 현실로 다가오는 신호**를 미리 정해 두는 일이다.
